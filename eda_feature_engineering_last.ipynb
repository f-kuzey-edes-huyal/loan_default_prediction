```python

# Assuming the target column is named 'target'
target_column = 'default'

# Calculate mutual information between the target and each feature
def calculate_mutual_information_with_target(df, target_column):
    features = df.drop(columns=[target_column]).columns
    mi_scores = {}

    for feature in features:
        mi_score = mutual_info_score(df[target_column].values, df[feature].values)
        mi_scores[feature] = mi_score

    return mi_scores

# Get mutual information scores
mi_scores = calculate_mutual_information_with_target(df_categorical, target_column)

# Convert to a DataFrame for easier visualization
mi_scores_df = pd.DataFrame(list(mi_scores.items()), columns=['Feature', 'Mutual Information'])

# Sort by mutual information score
mi_scores_df = mi_scores_df.sort_values(by='Mutual Information', ascending=False)

# Print the scores
print(mi_scores_df)

# Visualize the mutual information scores
plt.figure(figsize=(10, 6))
sns.barplot(x='Mutual Information', y='Feature', data=mi_scores_df, palette='pink')
plt.title("Mutual Information with Target")
plt.savefig('target_mutual_unformation.png', bbox_inches='tight')
plt.show()
```

```python
# Create the mutual information matrix
df_categorical = data_new.select_dtypes(include=object)
del df_categorical["loan_id"]
# Calculate mutual information for categorical variables
def calculate_mutual_information_categorical(df):
    mi_matrix = pd.DataFrame(index=df.columns, columns=df.columns, dtype=float)

    for feature_1 in df.columns:
        for feature_2 in df.columns:
            if feature_1 != feature_2:
                # Calculate mutual information score between the two features
                mi = mutual_info_score(df[feature_1], df[feature_2])
                mi_matrix.loc[feature_1, feature_2] = mi
            else:
                mi_matrix.loc[feature_1, feature_2] = np.nan  # Set diagonal to NaN for readability

    return mi_matrix

# Create the mutual information matrix
mi_matrix = calculate_mutual_information_categorical(df_categorical)
```

```python
mask2 = np.zeros_like(mi_matrix, dtype=bool)
mask2[np.triu_indices_from(mask2)] = True
np.fill_diagonal(mask2, False)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
plt.title("Mutual information matrix for categorical varables" , fontsize=14)

# Generate the heatmap including the mask
heatmap = sns.heatmap(mi_matrix,
                      annot=True,
                      annot_kws={"fontsize": 10},
                      fmt='.2f',
                      linewidths=0.5,
                      cmap='pink',
                      mask=mask2, # the mask has been included here
                      ax=ax)

# Display our plot
plt.savefig('categorical_mutual_information.png', bbox_inches='tight')

plt.show()
```

# Feature Engineering

```python
df_new_features = data.copy()
```

```python
# Income-to-Loan Ratio
df_new_features["income_to_loan_ratio"] = df_new_features["income"]/df_new_features["loan_amount"]
# Employment Stability
df_new_features["empl_sta"] = df_new_features["months_employed"] / df_new_features["age"]
# Loan Payment-to-Income Ratio
df_new_features["loan_pay_to_inc_ratio"] = df_new_features["loan_amount"]/(df_new_features["income"]*df_new_features["loan_term"])
#Credit Utilization Ratio
df_new_features["credit_util_ratio"] = df_new_features["loan_amount"]/ df_new_features["num_credit_lines"]
#Credit Age Factor
df_new_features["credit_age_factor"] = df_new_features["age"]/( df_new_features["num_credit_lines"])
#Interest Payment Burden: LoanAmount * InterestRate / 100
df_new_features["interest_payment_burden"] = df_new_features["loan_amount"]*(df_new_features["interest_rate"]/100)
#Income to Interest Ratio: Income / InterestRate
df_new_features["income_to_interest_ratio"] = df_new_features["income"] / df_new_features["interest_rate"]
#Credit Score and Income Interaction: CreditScore * Income
df_new_features["credit_income_inter"] = df_new_features["credit_score"]* df_new_features["income"]
#High DTI Flag: DTIRatio > Threshold
# Apply High DTI Flag
high_dti_threshold = 43
df_new_features["high_dti_flag"] = df_new_features["dtiratio"] > high_dti_threshold # Apply High DTI Flag 
#Marital Status & Dependents: MaritalStatus + "_" + HasDependents
df_new_features["marital_status_dependents"] = df_new_features["marital_status"] + "_" + df_new_features["has_dependents"]
#Employment Type and Loan Purpose: EmploymentType + "_" + LoanPurpose
df_new_features["employment_type_loan_purpose"] = df_new_features["employment_type"] + "_" + df_new_features["loan_purpose"]
```

# Removing Outliers 

```python
df_new_features.info()
```

```python
numerical_cols = df_new_features.select_dtypes(include = ["float64", "int64"]).columns
```

```python
def remove_outliers(df, cols):
    for col in cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5*IQR
        upper_bound = Q3 + 1.5*IQR
    
        df = df[(df[col] > lower_bound) & (df[col] <upper_bound)]
        return df
        
    
```

```python
#Remove outliers from the dataset
data_cleaned = remove_outliers(df_new_features, numerical_cols)
```
